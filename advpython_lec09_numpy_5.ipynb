```python
# 배열 연산 (array operation)

# 넘파이와 리스트의 배열 연산 실행 속도 비교
import random

n = 1000000

# 0~1사이의 실수를 uniform distribution에 따라 추출
a = [random.random() for i in range(n)]
b = [random.random() for i in range(n)]

arr1 = np.array(a)
arr2 = np.array(b)
print(arr1.shape)

%timeit [a[j] + b[j] for j in range(n)] # 리스트 연산
%timeit arr1 + arr2   # 넘파이 연산

%timeit sum(a)        # 파이썬 sum() 함수
%timeit np.sum(arr1)  # 넘파이 np.sum() 함수
print('='*80)

# 100번 반복하고, 3세트 실행할 때
%timeit -n 100 -r 3 [a[j] + b[j] for j in range(n)] # 리스트 연산
%timeit -n 100 -r 3 arr1 + arr2   # 넘파이 연산

%timeit -n 100 -r 3 sum(a)        # 파이썬 sum() 함수
%timeit -n 100 -r 3 np.sum(arr1)  # 넘파이 np.sum() 함수
```

```python
%%timeit
# 여러 줄의 코드의 시간을 측정할 때
# 주의1: %%timeit이 반드시 셀의 첫 번째 줄에서 시작해야 함
# 주의2: % 또는 %% 과 같은 매직 명령어는 Jupyter notebook이나 IPython 콘솔에서만 제공함

import random
n = 1000000
# 0~1사이의 실수를 uniform distribution에 따라 추출
a = [random.random() for i in range(n)]
b = [random.random() for i in range(n)]

result = []
for j in range(n):
    sum_val = a[j] + b[j]
    result.append(sum_val)
```

```python
# 일반적인 환경에서는..

import random
import timeit

def measure_list_operation():
    n = 1000000
    # 0~1사이의 실수를 uniform distribution에 따라 추출
    a = [random.random() for i in range(n)]
    b = [random.random() for i in range(n)]

    result = []
    for j in range(n):
        sum_val = a[j] + b[j]
        result.append(sum_val)
    return result

# number는 실행 횟수 조절 (%timeit 처럼 run, loops each가 없음)
execution_time = timeit.timeit(measure_list_operation, number=10) 
print(f"리스트 연산 실행 시간 (평균): {execution_time / 10:.6f} 초")
```

```python
# time.time() 함수 (매우 일반적이고 간단한 방법)
# 코드의 시작 시간과 끝 시간을 측정하고, 그 차이를 계산하여 실행 시간을 계산
# 비교적 긴 시간 동안 실행되는 코드 블록의 대략적인 실행 시간을 측정할 때 주로 사용
import time
import random

start_time = time.time()

n = 1000000
a = [random.random() for _ in range(n)]
b = [random.random() for _ in range(n)]

result = []
for j in range(n):
    sum_val = a[j] + b[j]
    result.append(sum_val)

end_time = time.time()

execution_time = end_time - start_time
print(f"실행 시간 (time.time()): {execution_time:.6f} 초")
```

```python
# 선형 대수 (linear algebra)

# np.linalg.solve( ) 함수
# Ax = b 문제 풀기
A = np.array([[1, 2, 3], [2, 0, 1], [1, -1, 1]])
c = np.array([[-2, 4, -1]])
b = c.T
x = np.linalg.solve(A, b)
print(x), print('='*30)

# 3*x0 + x1 = 9와 x0 + 2*x1 = 8의 연립 방정식 풀기
A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])
x = np.linalg.solve(A, b)
print(x)
```

```python
# 선형 대수 (linear algebra)

# dot product
# 1차원 벡터의 dot product
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.dot(a, b))     # 32
print('='*30)


# 2차원 이상의 배열의 dot product
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[4, 5], [6, 7]])
print(np.dot(arr1, arr2)), print('-'*30)
print(np.matmul(arr1, arr2)), print('='*30)

# 2차원 배열과 1차원 벡터의 dot product
c = np.array([2,1])
print(np.dot(arr1, c))
print(np.matmul(arr1, c))
```

```python
# Image Processing

import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('./cat.jpg')

print(f'shape: {img.shape}')
print(f'ndim: {img.ndim}')
print(f'min: {img.min()}')
print(f'max: {img.max()}')
print(f'mean: {img.mean()}')
print(f'dtype: {img.dtype}')

plt.figure()
plt.imshow(img)
plt.show()
```

```python
# Image Processing

def color2grayscale(color_img):
    converting_vec = [0.299, 0.587, 0.144]
    return np.dot(color_img, converting_vec)

gray_img = color2grayscale(img)

print(gray_img.shape)
print(gray_img.ndim)

plt.figure()
plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
# Image Processing

import scipy.ndimage
import imageio.v3 as iio

img_gray = iio.imread('./cat.jpg', mode='L')  # 흑백으로 읽힘

arr_flipud = np.flipud(img_gray)
arr_fliplr = np.fliplr(img_gray)
arr_rotate = scipy.ndimage.rotate(img_gray, 45)

plt.figure()
plt.subplot(2,2,1), plt.imshow(img_gray, cmap='gray')
plt.title('original gray image'), plt.axis('off')
plt.subplot(2,2,2), plt.imshow(arr_flipud, cmap='gray')
plt.title('upside down image'), plt.axis('off')
plt.subplot(2,2,3), plt.imshow(arr_fliplr, cmap='gray')
plt.title('left to right image'), plt.axis('off')
plt.subplot(2,2,4), plt.imshow(arr_rotate, cmap='gray')
plt.title('rotated image'), plt.axis('off')
plt.show()
```
