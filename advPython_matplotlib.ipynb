## Matplotlib

```python
# plt.plot()은 선 그래프를 그리는 함수

import matplotlib.pyplot as plt

#plt.plot(x축, y축)
plt.plot([1, 3, 7, 11, 14, 15]) # distance (y축)
#plt.plot([0, 1, 2, 3, 4, 5], [1, 3, 7, 11, 14, 15]) # distance (y축)
plt.xlabel('time[min]')
plt.ylabel('distance[km]')
# 이전에 생성된 그림들을 닫고, 다음 그림을 위한 figure를 새로 초기화
plt.show()


plt.plot([1,2,3,4,5], [120,200,160,350,110])
plt.xlabel('input')
plt.ylabel('output')
plt.title('Experiment Result')
plt.show()
```

```python
# 특성 지정 통한 선 스타일 변경 가능
# plt.xlabel(축이름), plt.ylabel(축이름), plt.title(제목),
# plt.xlim(범위), plt.ylim(범위) 등의 함수를 사용

import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')

# fontdict 매개변수는 텍스트의 다양한 글꼴 속성을 한 번에 지정할 수 있도록
# 딕셔너리 형태로 값을 받음
# fontdict={'fontsize': 10, 'fontweight':'bold', 'fontstyle':'italic','color':'red'}
plt.xlabel("time", fontdict={'fontsize': 10}) # x-axis label
plt.ylabel("amplitude", fontdict={'fontsize': 10}) # y-axis label
plt.title("plot subject", loc='center', fontdict={'fontsize': 10}) # title
#plt.title("plot subject", loc='center', fontsize=12, fontweight='bold') # title

plt.xlim(10, 50), plt.ylim(0, 20)
plt.show()
```

```python
# 먼저 matplotlib.pyplot 객체를 plt로 정의하고 불러온다.
# matplotlib에서 font_manager와 rc를 불러온다.

import matplotlib
import matplotlib.pyplot as plt
from matplotlib import font_manager

font_path="c:/Windows/Fonts/NGULIM.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rc('font',family=font_name)

y = (16,9,4,1,0,1,4,9,16)

plt.plot(y, 'b')
plt.title('파형')
plt.ylabel('출력')
plt.xlabel('입력')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) # 0부터 10까지 100개의 점 생성

# 첫 번째 선 (파란색, 실선, 약간 투명)
plt.plot(x, np.sin(x), color='blue', linestyle='-', linewidth=2, alpha=0.6, label='sin(x)')

# 두 번째 선 (빨간색, 점선, 마커 포함)
plt.plot(x, np.cos(x), color='red', linestyle='--', marker='o', markersize=2, label='cos(x)')

plt.xlabel("X-axis"), plt.ylabel("Y-axis")
plt.title("Multiple Lines with Custom Styles")
plt.legend() # 범례 표시
plt.grid(linestyle=':', alpha=0.6) # 그리드 추가
plt.show()
```

```python
# 여러 개의 선 그래프를 동시에 그리는 경우
# 각 선이 무슨 자료를 표시하는지를 보여주기 위해 legend 명령으로 범례를 추가할 수 있다.

from matplotlib import pyplot as plt

plt.plot([1,2,3,4], [1,4,9,10], 'rs--')
plt.plot([2,3,4,5],[5,6,7,9], 'b^-')
plt.xlabel('input')
plt.ylabel('output')
plt.title('Experiment Result')
plt.legend(['Dog', 'Cat'], loc='lower right')
plt.show()
```

```python
# 특성 지정 통한 선 스타일 변경 가능
# 포인트의 크기(s)와 색상, 투명도(alpha=)를 지정하는 예제

import matplotlib.pyplot as plt

x = range(0, 4)
y = [v**2 for v in x]
s = [100, 20, 30, 40]   # size

plt.scatter(x = x, y = y, s = s, c = 'red', alpha=0.6)
#plt.scatter(x, y, s, 'red', alpha=0.6)
plt.show()
```

```python
# subplot 케이스

import matplotlib.pyplot as plt

x = range(0, 100)
y1 = [v**2 for v in x]
y2 = [v**3 for v in x]

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.ylabel('v^2')

plt.subplot(2, 2, 4)
plt.plot(x, y2)
plt.ylabel('v^3')
plt.show()
```

```python
# 그래프 크기 케이스

import matplotlib.pyplot as plt

x = range(0, 100)
y1 = [v**2 for v in x]
y2 = [v**3 for v in x]

# figsize=(width, height)로 크기는 인치를 단위로 함
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.ylabel('v^2')

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.ylabel('v^3')
plt.grid()
plt.show()
```

```python
# 바 차트(bar chart)

import matplotlib.pyplot as plt

categories = ['Fruita', 'Apple', 'Banana', 'Orange']
sales = [150, 230, 180, 100]
colors = ['lightcoral', 'lightgreen', 'lightblue', 'lightgoldenrodyellow'] # 각 막대별 색상 지정

plt.figure(figsize=(9, 6))
plt.bar(categories, sales, width=0.6, color=colors, edgecolor='gray', linewidth=1.5)
plt.xlabel("Fruits")
plt.ylabel("Sales Volume")
plt.title("Sales Volume by Fruit Type")
plt.ylim(0, 250) # y축 범위 설정
plt.show()
```

```python
# 바 차트(bar chart): plt.barh()

import matplotlib.pyplot as plt

categories = ['Fruit', 'Apple', 'Banana', 'Orange']
sales = [150, 230, 180, 100]
colors = ['lightcoral', 'lightgreen', 'lightblue', 'lightgoldenrodyellow'] # 각 막대별 색상 지정

plt.figure(figsize=(9, 6))
plt.barh(categories, sales, align='center', color=colors, edgecolor='gray', linewidth=1.5)
plt.xlabel("Sales Volume")
plt.ylabel("Fruits")
plt.title("Sales Volume by Fruit Type")
plt.show()
```

```python
# 축 눈금 지정

import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 15, 7, 20, 12])

plt.figure(figsize=(7, 5))
plt.plot(x, y, marker='o')

# x축 눈금 위치와 레이블을 수동으로 설정
plt.xticks(ticks=[1, 2, 3, 4, 5], labels=['Jan', 'Feb', 'Mar', 'Apr', 'May'],
           rotation=45, fontsize=10, color='darkblue')

# y축 눈금 위치만 설정 (레이블은 자동으로 숫자)
plt.yticks(ticks=[5, 10, 15, 20], fontsize=10, color='darkgreen')

plt.xlabel("Month"), plt.ylabel("Sales"), plt.title("Monthly Sales Data")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout() # 레이블이 잘리지 않도록 조정
plt.show()
```

```python
# 객체지향 방식 subplot
# figure 함수를 통해 figure 객체를 먼저 만든 후 add_subplot 메소드를 통해
# 그리는 그래프 개수 만큼 subplot을 만든다.

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 4))
for subplot in range(1, 7):
    axis = fig.add_subplot(2, 3, subplot)
fig.tight_layout()


fig = plt.figure(figsize=(10, 3))
t = np.linspace(0.0, 1.0)
for subplot in range(1, 7):
    axis = fig.add_subplot(2, 3, subplot)
    axis.plot(t, np.cos(2*np.pi*t*subplot)+1)
    axis.set_xlabel('time[s]')
    axis.set_ylabel('voltage[mV]')
    axis.set_title('$\cos({} \pi t)$ +1'.format(subplot))
fig.tight_layout()
```

```python
# 그래프 안에 그래프

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10, 2)) # control figure size
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # define two axes
axis2 = fig.add_axes([0.45, 0.6, 0.1, 0.2])

x = np.arange(0, 180,1)
y = np.cos(2*x*np.pi/180)+1

axis1.plot(x, y) # add data to the both axes
axis1.set_xlabel('angle[degree]', fontsize=15)# define x-label for axis1
axis1.set_ylabel('signal[V]', fontsize=15)# define y-label for axis1

axis2.plot(x, y)
axis2.set_xbound(50, 70) # set the range of the second axis
axis2.set_ybound(0.3, 0.7)
axis2.set_xscale('log') # define x-scale for axis2

plt.show()
```
