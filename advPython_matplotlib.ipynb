```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) # 0 10 100  

#    (, ,  )
plt.plot(x, np.sin(x), color='blue', linestyle='-', linewidth=2, alpha=0.6, label='sin(x)')

#    (, ,  )
plt.plot(x, np.cos(x), color='red', linestyle='--', marker='o', markersize=2, label='cos(x)')

plt.xlabel("X-axis"), plt.ylabel("Y-axis")
plt.title("Multiple Lines with Custom Styles")
plt.legend() #  
plt.grid(linestyle=':', alpha=0.6) #  
plt.show()
```

```python
#       
#  (s) , (alpha=)  

import matplotlib.pyplot as plt

x = range(0, 4)
y = [v**2 for v in x]
s = [100, 20, 30, 40]   # size

plt.scatter(x = x, y = y, s = s, c = 'red', alpha=0.6)
#plt.scatter(x, y, s, 'red', alpha=0.6)
plt.show()
```

```python
# subplot 

import matplotlib.pyplot as plt

x = range(0, 100)
y1 = [v**2 for v in x]
y2 = [v**3 for v in x]

plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.ylabel('v^2')

plt.subplot(2, 2, 4)
plt.plot(x, y2)
plt.ylabel('v^3')
plt.show()
```

```python
#   

import matplotlib.pyplot as plt

x = range(0, 100)
y1 = [v**2 for v in x]
y2 = [v**3 for v in x]

# figsize=(width, height)    
plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.ylabel('v^2')

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.ylabel('v^3')
plt.grid()
plt.show()
```

```python
#  (bar chart)

import matplotlib.pyplot as plt

categories = ['Fruita', 'Apple', 'Banana', 'Orange']
sales = [150, 230, 180, 100]
colors = ['lightcoral', 'lightgreen', 'lightblue', 'lightgoldenrodyellow'] #    

plt.figure(figsize=(9, 6))
plt.bar(categories, sales, width=0.6, color=colors, edgecolor='gray', linewidth=1.5)
plt.xlabel("Fruits")
plt.ylabel("Sales Volume")
plt.title("Sales Volume by Fruit Type")
plt.ylim(0, 250) # y  
plt.show()
```

```python
#  (bar chart): plt.barh()

import matplotlib.pyplot as plt

categories = ['Fruit', 'Apple', 'Banana', 'Orange']
sales = [150, 230, 180, 100]
colors = ['lightcoral', 'lightgreen', 'lightblue', 'lightgoldenrodyellow'] #    

plt.figure(figsize=(9, 6))
plt.barh(categories, sales, align='center', color=colors, edgecolor='gray', linewidth=1.5)
plt.xlabel("Sales Volume")
plt.ylabel("Fruits")
plt.title("Sales Volume by Fruit Type")
plt.show()
```

```python
#   

import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 15, 7, 20, 12])

plt.figure(figsize=(7, 5))
plt.plot(x, y, marker='o')

# x     
plt.xticks(ticks=[1, 2, 3, 4, 5], labels=['Jan', 'Feb', 'Mar', 'Apr', 'May'],
           rotation=45, fontsize=10, color='darkblue')

# y    (  )
plt.yticks(ticks=[5, 10, 15, 20], fontsize=10, color='darkgreen')

plt.xlabel("Month"), plt.ylabel("Sales"), plt.title("Monthly Sales Data")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout() #    
plt.show()
```

```python
#   subplot
# figure   figure     add_subplot  
#     subplot .

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 4))
for subplot in range(1, 7):
    axis = fig.add_subplot(2, 3, subplot)
fig.tight_layout()


fig = plt.figure(figsize=(10, 3))
t = np.linspace(0.0, 1.0)
for subplot in range(1, 7):
    axis = fig.add_subplot(2, 3, subplot)
    axis.plot(t, np.cos(2*np.pi*t*subplot)+1)
    axis.set_xlabel('time[s]')
    axis.set_ylabel('voltage[mV]')
    axis.set_title('$\cos({} \pi t)$ +1'.format(subplot))
fig.tight_layout()
```

```python
#   

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10, 2)) # control figure size
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # define two axes
axis2 = fig.add_axes([0.45, 0.6, 0.1, 0.2])

x = np.arange(0, 180,1)
y = np.cos(2*x*np.pi/180)+1

axis1.plot(x, y) # add data to the both axes
axis1.set_xlabel('angle[degree]', fontsize=15)# define x-label for axis1
axis1.set_ylabel('signal[V]', fontsize=15)# define y-label for axis1

axis2.plot(x, y)
axis2.set_xbound(50, 70) # set the range of the second axis
axis2.set_ybound(0.3, 0.7)
axis2.set_xscale('log') # define x-scale for axis2

plt.show()
```
